## Cleaning the tweet text by Natural Language Processing using the Natural Language Toolkit or NLTK 

```python
tweets = list(df['text'])
tweets[5]
```

### We have to remove the usernames that start with '@', links, punctuations, and stopwords. We will also get the root words using lemmatizer.

```python
def strip_all_entities(text):
    words = []
    entity_prefixes = ['@','#','\\']

    for word in text.split():
        word = word.strip()
        if word:
            if word[0] not in entity_prefixes:
                words.append(word)
    return ' '.join(words)


for i in range(0,len(tweets)):
    tweets[i] = tweets[i].lower()
    tweets[i] = strip_all_entities(tweets[i])

tweets[5]
```

```python
def remove_links(text):
    words = []
    for word in text.split():
        if not 'https' in word:
            words.append(word)
    return ' '.join(words)


for i in range(0,len(tweets)):
    tweets[i] = remove_links(tweets[i])
    tweets[i] = tweets[i].replace("[^a-zA-Z#]"," ")
tweets[0:5]
```

```python
def remove_punc(text):
    words = nltk.word_tokenize(text)
    words=[word for word in words if word.isalpha()]
    return ' '.join(words)

for i in range(0,len(tweets)):
    tweets[i] = remove_punc(tweets[i])

tweets[0:10]
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = nltk.word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    return ' '.join(words)

for i in range(0,len(tweets)):
    tweets[i] = remove_stopwords(tweets[i])

tweets[0:10]
```

```python
from nltk.stem import WordNetLemmatizer

lemm = WordNetLemmatizer()

def get_root_words(text):
    words = nltk.word_tokenize(text)
    words = [lemm.lemmatize(word) for word in words]
    return " ".join(words)

for i in range(0,len(tweets)):
    tweets[i] = get_root_words(tweets[i])

tweets[0:10]
```
